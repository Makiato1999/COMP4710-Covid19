```python
import pandas as pd
import datetime as dt
from pandas.tseries.offsets import MonthEnd
```

```python
def readCSV(inputCSV):
    df = pd.read_csv(inputCSV)
    #display(df)
    
readCSV("Merged_Datasets(date)/Philippine_Monthly.csv")
```

```python
def convert_to_diff_dataset(input_csv , output_csv):
    input_df = pd.read_csv(input_csv)
    df = input_df.copy()   
    df["canada_unemployment_rates"] = df["canada_unemployment_rates"].diff()
    df["unemployment_rates"] = df["unemployment_rates"].diff()
    df["canada_inflation_rate"] = df["canada_inflation_rate"].diff()
    df["inflation_rate"] = df["inflation_rate"].diff()
    df["canada_total_cases"] = df["canada_total_cases"].pct_change()
    df["canada_total_tests"] = df["canada_total_tests"].pct_change()
    df["canada_people_fully_vaccinated"] = df["canada_people_fully_vaccinated"].pct_change()
    df["canada_total_deaths"] = df["canada_total_deaths"].pct_change()
    df["canada_reproduction_rate"] = df["canada_reproduction_rate"].diff()
    df["total_cases"] = df["total_cases"].pct_change()
    df["total_tests"] = df["total_tests"].pct_change()
    df["people_fully_vaccinated"] = df["people_fully_vaccinated"].pct_change()
    df["total_deaths"] = df["total_deaths"].pct_change()
    df["reproduction_rate"] = df["reproduction_rate"].diff()
    df.reset_index(drop=True, inplace=True)
    df.to_csv(output_csv , index=True)
    return(df)


France_df = convert_to_diff_dataset("Merged_Datasets(date)/France_Monthly.csv" , "Merged_Datasets(Diff)/France_Monthly_Diff.csv")
Philippine_df = convert_to_diff_dataset("Merged_Datasets(date)/Philippine_Monthly.csv" , "Merged_Datasets(Diff)/Philippine_Monthly_Diff.csv")
SouthKorea_df = convert_to_diff_dataset("Merged_Datasets(date)/SouthKorea_Monthly.csv" , "Merged_Datasets(Diff)/SouthKorea_Monthly_Diff.csv")
China_df = convert_to_diff_dataset("Merged_Datasets(date)/China_Monthly.csv" , "Merged_Datasets(Diff)/China_Monthly_Diff.csv")
India_df = convert_to_diff_dataset("Merged_Datasets(date)/India_Monthly.csv" , "Merged_Datasets(Diff)/India_Monthly_Diff.csv")
France_df.head()
```

```python
def unemploy_rates_cons(canada_unem_rates , unem_rates):
    df_1 = pd.DataFrame()
    df_1.insert(0, "canada_unemployment_rates_value", canada_unem_rates)
    df_1["canada_unemployment_rates"] = False
    df_1.loc[df_1['canada_unemployment_rates_value'] > 0, ["canada_unemployment_rates"]] = True
    df_1.pop("canada_unemployment_rates_value")
    
    df_2 = pd.DataFrame()
    df_2.insert(0, "unemployment_rates_value", unem_rates)
    df_2["unemployment_rates"] = False
    df_2.loc[df_2['unemployment_rates_value'] > 0, ["unemployment_rates"]] = True
    df_1.insert(1, "unemployment_rates", df_2["unemployment_rates"])   
    
    #df_1.insert(3, "unemployment_rates_value", df_2["unemployment_rates_value"])   
    #display(df_1)
    return (df_1)

def inflat_rates_cons(canada_inflat_rates , inflat_rates):
    df_1 = pd.DataFrame()
    df_1.insert(0, "canada_inflation_rate_value", canada_inflat_rates)
    df_1["canada_inflation_rate"] = False
    df_1.loc[df_1['canada_inflation_rate_value'] > 0, ["canada_inflation_rate"]] = True
    df_1.pop("canada_inflation_rate_value")
    
    df_2 = pd.DataFrame()
    df_2.insert(0, "inflation_rate_value", inflat_rates)
    df_2["inflation_rate"] = False
    df_2.loc[df_2['inflation_rate_value'] > 0, ["inflation_rate"]] = True
    df_1.insert(1, "inflation_rate", df_2["inflation_rate"])   
      
    #display(df_1)
    return (df_1)

def covid_info_cons1(can_total_cases , total_cases , can_total_tests , total_tests , can_ppl_vac , ppl_vac):
    df_1 = pd.DataFrame()
    df_1.insert(0, "canada_total_cases_value", can_total_cases)
    df_1["canada_total_cases"] = False
    df_1.loc[df_1['canada_total_cases_value'] > 0, ["canada_total_cases"]] = True
    df_1.pop("canada_total_cases_value")
    
    df_2 = pd.DataFrame()
    df_2.insert(0, "total_cases_value", total_cases)
    df_2["total_cases"] = False
    df_2.loc[df_2['total_cases_value'] > 0, ["total_cases"]] = True
    df_1.insert(1, "total_cases", df_2["total_cases"])   
    
    df_3 = pd.DataFrame()
    df_3 = df_3.astype(bool)  
    df_3.insert(0, "canada_total_tests_value", can_total_tests)
    df_3["canada_total_tests"] = False
    df_3.loc[df_3['canada_total_tests_value'] > 0, ["canada_total_tests"]] = True
    df_1.insert(1, "canada_total_tests", df_3["canada_total_tests"])   
    
    df_4 = pd.DataFrame() 
    df_4.insert(0, "total_tests_value", total_tests)
    df_4["total_tests"] = False
    df_4.loc[df_4['total_tests_value'] > 0, ["total_tests"]] = True
    df_1.insert(1, "total_tests", df_4["total_tests"])   
    
    df_5 = pd.DataFrame() 
    df_5.insert(0, "canada_people_fully_vaccinated_value", can_ppl_vac)
    df_5["canada_people_fully_vaccinated"] = False
    df_5.loc[df_5['canada_people_fully_vaccinated_value'] > 0, ["canada_people_fully_vaccinated"]] = True
    df_1.insert(1, "canada_people_fully_vaccinated", df_5["canada_people_fully_vaccinated"])  
    
    df_6 = pd.DataFrame()
    df_6.insert(0, "people_fully_vaccinated_value", can_ppl_vac)
    df_6["people_fully_vaccinated"] = False
    df_6.loc[df_6['people_fully_vaccinated_value'] > 0, ["people_fully_vaccinated"]] = True
    df_1.insert(1, "people_fully_vaccinated", df_6["people_fully_vaccinated"])
    
    #display(df_1)
    return (df_1)

def covid_info_cons2(can_total_death , total_death , can_repro_rates , repro_rates):
    df_1 = pd.DataFrame() 
    df_1.insert(0, "canada_total_deaths_value", can_total_death)
    df_1["canada_total_deaths"] = False
    df_1.loc[df_1['canada_total_deaths_value'] > 0, ["canada_total_deaths"]] = True
    df_1.pop("canada_total_deaths_value")
    
    df_2 = pd.DataFrame()  
    df_2.insert(0, "total_deaths_value", total_death)
    df_2["total_deaths"] = False
    df_2.loc[df_2['total_deaths_value'] > 0, ["total_deaths"]] = True
    df_1.insert(1, "total_deaths", df_2["total_deaths"]) 
    
    df_3 = pd.DataFrame() 
    df_3.insert(0, "canada_reproduction_rate_value", can_repro_rates)
    df_3["canada_reproduction_rate"] = False
    df_3.loc[df_3['canada_reproduction_rate_value'] > 0, ["canada_reproduction_rate"]] = True
    df_1.insert(1, "canada_reproduction_rate", df_3["canada_reproduction_rate"])   
    
    df_4 = pd.DataFrame() 
    df_4.insert(0, "reproduction_rate_value", repro_rates)
    df_4["reproduction_rate"] = False
    df_4.loc[df_4['reproduction_rate_value'] > 0, ["reproduction_rate"]] = True
    df_1.insert(1, "reproduction_rate", df_4["reproduction_rate"])   

    #display(df_1)
    return (df_1)

def convert_to_transaction(inputCSV , outputCSV):
    input_df = pd.read_csv(inputCSV)
    input_df = input_df.iloc[1: , :]  #delete first row
    df = pd.DataFrame()

    unemploy_df = unemploy_rates_cons(input_df["canada_unemployment_rates"] , input_df["unemployment_rates"] )
    df["canada_unemployment_rates"] = unemploy_df["canada_unemployment_rates"]
    df["unemployment_rates"] = unemploy_df["unemployment_rates"]
    
    inflat_df = inflat_rates_cons(input_df["canada_inflation_rate"] , input_df["inflation_rate"])
    df["canada_inflation_rate"] = inflat_df["canada_inflation_rate"]
    df["inflation_rate"] = inflat_df["inflation_rate"]
    
    covid_info1_df = covid_info_cons1(input_df["canada_total_cases"] , input_df["total_cases"] , input_df["canada_total_tests"] , input_df["total_tests"] , input_df["canada_people_fully_vaccinated"] , input_df["people_fully_vaccinated"])
    df["canada_total_cases"] = covid_info1_df["canada_total_cases"]
    df["total_cases"] = covid_info1_df["total_cases"]
    df["canada_total_tests"] = covid_info1_df["canada_total_tests"]
    df["total_tests"] = covid_info1_df["total_tests"]
    df["canada_people_fully_vaccinated"] = covid_info1_df["canada_people_fully_vaccinated"]
    df["people_fully_vaccinated"] = covid_info1_df["people_fully_vaccinated"]
    
    covid_info2_df = covid_info_cons2(input_df["canada_total_deaths"] , input_df["total_deaths"] , input_df["canada_reproduction_rate"] , input_df["reproduction_rate"])
    df["canada_total_deaths"] = covid_info2_df["canada_total_deaths"]
    df["total_deaths"] = covid_info2_df["total_deaths"]
    df["canada_reproduction_rate"] = covid_info2_df["canada_reproduction_rate"]
    df["reproduction_rate"] = covid_info2_df["reproduction_rate"]
    
    #df["Transaction ID"] = df.index
    #first_column = df.pop('Transaction ID')
    #df.insert(0, 'Transaction ID', first_column)    
    #display(df)
    
    df.reset_index(drop=True, inplace=True)
    df.to_csv(outputCSV , index=False)
    return(df)

convert_to_transaction("Merged_Datasets(Diff)/India_Monthly_Diff.csv" , "Transaction_Datasets/India_Monthly_Diff.csv")
convert_to_transaction("Merged_Datasets(Diff)/France_Monthly_Diff.csv" , "Transaction_Datasets/France_Monthly_Diff.csv")
convert_to_transaction("Merged_Datasets(Diff)/SouthKorea_Monthly_Diff.csv" , "Transaction_Datasets/SouthKorea_Monthly_Diff.csv")
convert_to_transaction("Merged_Datasets(Diff)/Philippine_Monthly_Diff.csv" , "Transaction_Datasets/Philippine_Monthly_Diff.csv")
convert_to_transaction("Merged_Datasets(Diff)/China_Monthly_Diff.csv" , "Transaction_Datasets/China_Monthly_Diff.csv")
```

```python
def unemploy_rates_cons_2(canada_unem_rates , unem_rates):
    df_1 = pd.DataFrame()
    df_1.insert(0, "canada_unemployment_rates_value", canada_unem_rates)
    df_1["canada_unemployment_rates"] = False
    df_1.loc[df_1['canada_unemployment_rates_value'] > 0.1, ["canada_unemployment_rates"]] = True
    df_1.pop("canada_unemployment_rates_value")
    
    df_2 = pd.DataFrame()
    df_2.insert(0, "unemployment_rates_value", unem_rates)
    df_2["unemployment_rates"] = False
    df_2.loc[df_2['unemployment_rates_value'] > 0.1, ["unemployment_rates"]] = True
    df_1.insert(1, "unemployment_rates", df_2["unemployment_rates"])   
    
    return (df_1)

def inflat_rates_cons_2(canada_inflat_rates , inflat_rates):
    df_1 = pd.DataFrame()
    df_1.insert(0, "canada_inflation_rate_value", canada_inflat_rates)
    df_1["canada_inflation_rate"] = False
    df_1.loc[df_1['canada_inflation_rate_value'] > 0.1 , ["canada_inflation_rate"]] = True
    df_1.pop("canada_inflation_rate_value")
    
    df_2 = pd.DataFrame()
    df_2.insert(0, "inflation_rate_value", inflat_rates)
    df_2["inflation_rate"] = False
    df_2.loc[df_2['inflation_rate_value'] > 0.1, ["inflation_rate"]] = True
    df_1.insert(1, "inflation_rate", df_2["inflation_rate"])   
      
    return (df_1)

def covid_info_cons1_2(can_total_cases , total_cases , can_total_tests , total_tests , can_ppl_vac , ppl_vac):
    df_1 = pd.DataFrame()
    df_1.insert(0, "canada_total_cases_value", can_total_cases)
    df_1["canada_total_cases"] = False
    df_1.loc[df_1['canada_total_cases_value'] > 0.1, ["canada_total_cases"]] = True
    df_1.pop("canada_total_cases_value")
    
    df_2 = pd.DataFrame()
    df_2.insert(0, "total_cases_value", total_cases)
    df_2["total_cases"] = False
    df_2.loc[df_2['total_cases_value'] > 0.1, ["total_cases"]] = True
    df_1.insert(1, "total_cases", df_2["total_cases"])   
    
    df_3 = pd.DataFrame()
    df_3 = df_3.astype(bool)  
    df_3.insert(0, "canada_total_tests_value", can_total_tests)
    df_3["canada_total_tests"] = False
    df_3.loc[df_3['canada_total_tests_value'] > 0.1, ["canada_total_tests"]] = True
    df_1.insert(1, "canada_total_tests", df_3["canada_total_tests"])   
    
    df_4 = pd.DataFrame() 
    df_4.insert(0, "total_tests_value", total_tests)
    df_4["total_tests"] = False
    df_4.loc[df_4['total_tests_value'] > 0.1, ["total_tests"]] = True
    df_1.insert(1, "total_tests", df_4["total_tests"])   
    
    df_5 = pd.DataFrame() 
    df_5.insert(0, "canada_people_fully_vaccinated_value", can_ppl_vac)
    df_5["canada_people_fully_vaccinated"] = False
    df_5.loc[df_5['canada_people_fully_vaccinated_value'] > 0.1, ["canada_people_fully_vaccinated"]] = True
    df_1.insert(1, "canada_people_fully_vaccinated", df_5["canada_people_fully_vaccinated"])  
    
    df_6 = pd.DataFrame()
    df_6.insert(0, "people_fully_vaccinated_value", can_ppl_vac)
    df_6["people_fully_vaccinated"] = False
    df_6.loc[df_6['people_fully_vaccinated_value'] > 0.1, ["people_fully_vaccinated"]] = True
    df_1.insert(1, "people_fully_vaccinated", df_6["people_fully_vaccinated"])
    
    return (df_1)

def covid_info_cons2_2(can_total_death , total_death , can_repro_rates , repro_rates):
    df_1 = pd.DataFrame() 
    df_1.insert(0, "canada_total_deaths_value", can_total_death)
    df_1["canada_total_deaths"] = False
    df_1.loc[df_1['canada_total_deaths_value'] > 0.1, ["canada_total_deaths"]] = True
    df_1.pop("canada_total_deaths_value")
    
    df_2 = pd.DataFrame()  
    df_2.insert(0, "total_deaths_value", total_death)
    df_2["total_deaths"] = False
    df_2.loc[df_2['total_deaths_value'] > 0.1, ["total_deaths"]] = True
    df_1.insert(1, "total_deaths", df_2["total_deaths"]) 
    
    df_3 = pd.DataFrame() 
    df_3.insert(0, "canada_reproduction_rate_value", can_repro_rates)
    df_3["canada_reproduction_rate"] = False
    df_3.loc[df_3['canada_reproduction_rate_value'] > 0.1, ["canada_reproduction_rate"]] = True
    df_1.insert(1, "canada_reproduction_rate", df_3["canada_reproduction_rate"])   
    
    df_4 = pd.DataFrame() 
    df_4.insert(0, "reproduction_rate_value", repro_rates)
    df_4["reproduction_rate"] = False
    df_4.loc[df_4['reproduction_rate_value'] > 0.1, ["reproduction_rate"]] = True
    df_1.insert(1, "reproduction_rate", df_4["reproduction_rate"])   

    return (df_1)

def convert_to_transaction_2(inputCSV , outputCSV):
    input_df = pd.read_csv(inputCSV)
    input_df = input_df.iloc[1: , :]  #delete first row
    df = pd.DataFrame()

    unemploy_df = unemploy_rates_cons_2(input_df["canada_unemployment_rates"] , input_df["unemployment_rates"] )
    df["canada_unemployment_rates"] = unemploy_df["canada_unemployment_rates"]
    df["unemployment_rates"] = unemploy_df["unemployment_rates"]
    
    inflat_df = inflat_rates_cons_2(input_df["canada_inflation_rate"] , input_df["inflation_rate"])
    df["canada_inflation_rate"] = inflat_df["canada_inflation_rate"]
    df["inflation_rate"] = inflat_df["inflation_rate"]
    
    covid_info1_df = covid_info_cons1_2(input_df["canada_total_cases"] , input_df["total_cases"] , input_df["canada_total_tests"] , input_df["total_tests"] , input_df["canada_people_fully_vaccinated"] , input_df["people_fully_vaccinated"])
    df["canada_total_cases"] = covid_info1_df["canada_total_cases"]
    df["total_cases"] = covid_info1_df["total_cases"]
    df["canada_total_tests"] = covid_info1_df["canada_total_tests"]
    df["total_tests"] = covid_info1_df["total_tests"]
    df["canada_people_fully_vaccinated"] = covid_info1_df["canada_people_fully_vaccinated"]
    df["people_fully_vaccinated"] = covid_info1_df["people_fully_vaccinated"]
    
    covid_info2_df = covid_info_cons2_2(input_df["canada_total_deaths"] , input_df["total_deaths"] , input_df["canada_reproduction_rate"] , input_df["reproduction_rate"])
    df["canada_total_deaths"] = covid_info2_df["canada_total_deaths"]
    df["total_deaths"] = covid_info2_df["total_deaths"]
    df["canada_reproduction_rate"] = covid_info2_df["canada_reproduction_rate"]
    df["reproduction_rate"] = covid_info2_df["reproduction_rate"]
    
    df.reset_index(drop=True, inplace=True)
    df.to_csv(outputCSV , index=False)
    return(df)

convert_to_transaction_2("Merged_Datasets(Diff)/India_Monthly_Diff.csv" , "Transaction_Datasets(2)/India_Monthly_Diff.csv")
convert_to_transaction_2("Merged_Datasets(Diff)/France_Monthly_Diff.csv" , "Transaction_Datasets(2)/France_Monthly_Diff.csv")
convert_to_transaction_2("Merged_Datasets(Diff)/SouthKorea_Monthly_Diff.csv" , "Transaction_Datasets(2)/SouthKorea_Monthly_Diff.csv")
convert_to_transaction_2("Merged_Datasets(Diff)/Philippine_Monthly_Diff.csv" , "Transaction_Datasets(2)/Philippine_Monthly_Diff.csv")
convert_to_transaction_2("Merged_Datasets(Diff)/China_Monthly_Diff.csv" , "Transaction_Datasets(2)/China_Monthly_Diff.csv")
```
