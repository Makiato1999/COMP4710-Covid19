```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

## COVID-19 total cases

```python
plt.rcParams["figure.figsize"] = [8.00, 5.50]
plt.rcParams["figure.autolayout"] = True


F_headers = ['canada_total_cases','total_cases']
headers = ['total_cases']

F_df = pd.read_csv('Merged_Datasets/France_Monthly.csv', usecols=F_headers)
I_df = pd.read_csv('Merged_Datasets/India_Monthly.csv', usecols=headers)
P_df = pd.read_csv('Merged_Datasets/Philippine_Monthly.csv', usecols=headers)
S_df = pd.read_csv('Merged_Datasets/SouthKorea_Monthly.csv', usecols=headers)
C_df = pd.read_csv('Merged_Datasets/China_Monthly.csv', usecols=headers)

ax = F_df.plot()
ax = I_df.plot(ax=ax)
ax = P_df.plot(ax=ax)
ax = S_df.plot(ax=ax)
C_df.plot(ax=ax)

plt.legend(['Canada','France',
           'India','Philippine',
            'SouthKorea','China'])
plt.title(r'COVID-19 total cases')
fig.tight_layout()
plt.show()
```

## Country Diff Datasets

```python
plt.rcParams["figure.figsize"] = [8.00, 6.50]
plt.rcParams["figure.autolayout"] = True


F_headers = ['canada_inflation_rate','inflation_rate']
headers = ['inflation_rate']

F_df = pd.read_csv('Merged_Datasets(Diff)/France_Monthly_Diff.csv', usecols=F_headers)
I_df = pd.read_csv('Merged_Datasets(Diff)/India_Monthly_Diff.csv', usecols=headers)
P_df = pd.read_csv('Merged_Datasets(Diff)/Philippine_Monthly_Diff.csv', usecols=headers)
S_df = pd.read_csv('Merged_Datasets(Diff)/SouthKorea_Monthly_Diff.csv', usecols=headers)
C_df = pd.read_csv('Merged_Datasets(Diff)/China_Monthly_Diff.csv', usecols=headers)

ax = F_df.plot()
ax = I_df.plot(ax=ax)
ax = P_df.plot(ax=ax)
ax = S_df.plot(ax=ax)
C_df.plot(ax=ax)

plt.legend(['Canada','France',
           'India','Philippine',
            'SouthKorea','China'])
plt.title(r'Inflation Rates Difference')
#fig.tight_layout()
plt.show()
```
